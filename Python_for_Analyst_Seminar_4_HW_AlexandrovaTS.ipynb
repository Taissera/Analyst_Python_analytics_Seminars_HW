2.2 Изучите распределение этажей домов

2.3 Изучите распределение состояния домов

```python
#2.1 Изучите распределение домов от наличия вида на набережную
data = df['waterfront'].value_counts()

names = data.index
values = data.value_counts

data
```

```python
plt.figure(figsize=(6,4))
plt.pie(data, autopct='%1.1f%%')
plt.title('Распределение (вид на набережную)')
plt.legend(data.index);
```

```python
#2.2 Изучите распределение этажей домов
data = df['floors'].value_counts()

names = data.indexvalues = data.values

data
```

```python
plt.figure(figsize=(6,4))
plt.pie(data, autopct='%1.1f%%', labels=names)
plt.title('Распределение этажей домов')
plt.legend(data.index);
```

```python
# 2.3 Изучите распределение состояния домов

data = df['condition'].value_counts()

names = data.indexvalues = data.values

data
```

```python
plt.figure(figsize=(6,4))
plt.pie(data, autopct='%1.1f%%', labels=names)
plt.title('Распределение состояния домов')
plt.legend(data.index);
```

## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5) #.astype('str')

plt.xlabel('price')
plt.ylabel('grade')
plt.title('Распределение цены по качеству конструкции и дизайна');
```

**^ Распределение цены по качеству конструкции и дизайна ^**

Получаем, что чем выше качество конструкции и дизайна, тем цена выше цена
При этом уровень 1-7 находятся практически в одном ценовом диапазоне. 
Далее по каждому уровню качества пиапазон цен увеличивается значительнее. 

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('waterfront')
plt.title('Распределение цены по наличию набережной');
```

**^ Распределение цены по наличию набережной ^**

Получаем чёткое разделение диапазона цен при наличии/отсутствиии набережной.
При наличии набережной цена дороже.

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('view')
plt.title('Распределение цены по виду');
```

**^ Распределение цены по виду ^**

Получаем, что чем лучше вид, тем цена выше цена. При этом уровень 0-3 находятся практически в одном ценовом диапазоне. 

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['bedrooms'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('bedrooms')
plt.title('Распределение цены по кол-ву комнат');
```

```python
plt.figure(figsize=(16,8))

sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('floors')
plt.title('Распределение цены по этажности');
```

**^ Распределение цены по этажности ^**

Получаем, что большого разброса цен в зависимости от этажности нет. 

При этом у этажности 2.5 диапазон цен больше относительно других значений.
